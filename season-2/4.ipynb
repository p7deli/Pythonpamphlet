### نحوه نمایش خروجی ها

## مدل اول

```python
# نمونه اول نمونه نسبتا قدیمی است
# این مورد شامل یک فرمول هست که شما برای استفاده از آن حتما باید تایپ(نوع) آن مقدار را مشخص کنید بقیه موارد اختیاری است


# [(key)] [flag] [w] [.p] type


# نمونه مثال برای TYPE
# ----------------------------------------

# %i برای عدد های صحیح
print("x is: %i\ny is: %i " %(25, 24))
# output: x is: 25
#         y is: 24 

# %c  برای تک کاراکترها یا عدد
print("%c or %c" % (256, 'p'))
# output: Ā or p

# %s برای نشان دادن رشته ها
print("name id %s and family %s" %('poria', 'deli'))
# output: name id poria and family deli

# %f برای اعداد اعشاری
print("Average: %f" %(14.68))
# output: Average: 14.680000


# نمونه مثال برای [.p]
# ------------------------------------------

# %.2f نمونه دیگر برای نشان دادن اعداد اعشاری و مشخص کردن تعداد اعشار آن
print("Average: %.2f" %(14.68))
# output: Average: 14.68


# نمونه مثال برای [w]
# ------------------------------------------

# برای فاصله انداخل از سمت راست یا چپ استفاده میشود
print("%9.f3" %(254.45))
# output:       2543


# نمونه مثال برای [flag]
# ------------------------------------------

# برای نشان دادن علامت مثبت یا منفی استفاده میشود
print("%+.2f\n%-.2f" %(21.54, -36.15))
# output: +21.54
#         -36.15

# نمونه مثال برای [(key)]
# ------------------------------------------

# برای استفاده از دیکشنری ها
d = {"a": 45, "b": 12.5}
print("%(a)f ---- %(b)i" %(d))
# output: 45.000000 ---- 12
```

## مدل دوم

```python
# در این نوع مدل نمایش کلی زیاد فرقی ندارد ولی ساختار یکم متفاوت است به طور کلی به شکل زیر است
x = 25
y = 15.6
print("x is {}\ny is {}\nz is {}".format(x, y, "poria"))
# output: x is 25
#         y is 15.6
#         z is poria

# همین طور شما با شماره گذاری هم میتوانید ترتیب آن ها را مشخص کنید ولی دقت کنید از 0 باید شروع کنید
print("x is {0}\ny is {2}\nz is {1}".format(x, y, "poria"))
# output: x is 25
#         y is poria
#         z is 15.6

# برای استفاده از دیکشنری ها باید به صورت زیر عمل کنیم
d = {"a": 25, "b": 17, "c": 36.25}
print("a is {a} b is {b} c is {c}".format(**d))
# output: a is 25 b is 17 c is 36.25

# استفاده از کاراکتر های رشته ها به ترتیب
print("0:{0} 1:{1} 2:{2}".format(*"por"))
# output: 0:p 1:o 2:r

# استفاده از عناصر لیست
print("0:{0} 1:{1} 2:{2}".format(*[1, 2, 8]))
# output: 0:1 1:2 2:8

# ---------------- فرمول از موارد اضافی ---------------
# "{[field_name] ["!" conversion] [":" format_spec]}"

# [field_name] همون شماره گذاری ها هستند و اختیاری است
print("{}{}{}".format(2, 5, "poria"))
# output: 25poria
print("{1}{0}{2}".format(8, 5, "poria"))
# output: 58poria


# ["!" conversion] با این علامت میتوانیم نوع مقدار را مشخص کنیم
# !s = str ------------ !a = acci cod ----------- !r = reper and ...
print("{!s} {!a} {!r}".format("poria", "+", "ali"))
# output: poria '+' 'ali'

# [":" format_spec] فرمت را مشخص میکنیم
# [[Fill]align] [sign] [#] [0] [width] [arouping_option] [.precision][type]
# همه ی این موارد اختیاری و بعد از دونقطه استفاده میشود
# ----------------------------------
# [type] 
# به جای i 
# از d استفاده میکنیم
print("{:f} {:d}".format(25., 25))
# output: 25.4748 25
# ----------------------------------
# [.precision]
print("{:.2f}".format(25.26580))
# output: 25.26
# ----------------------------------
# [arouping_option]
print("{:,}".format(25665487))
# output: 25,665,487

print("{:_}".format(25665487))
# output: 25_665_487
# ----------------------------------
# [width]
print("{:5}".format(12))
# output:   12
# ----------------------------------
# [0]
print("{:05}*".format(12))
# output:00012*
# ----------------------------------
# [#]
print("{:#b} {:#o} {:#x}".format(12, 14, 16))
# output:0b1100 0o16 0x10
# ----------------------------------
# [sign]
print("{:+} {:-}".format(12, -12))
# output:+12 -12
# ----------------------------------
# [Fill]align]
print("{:<15}*".format(14))
# output:14             *
print("{:>15}*".format(14))
# output:             14*
print("*{:^15}*".format(14))
# output:*      14       *
print("{:0>15}*".format(14))
# output:000000000000014*
```
