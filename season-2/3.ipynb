### حرکت روی ایندکس های رشته ها

```python
# شماره گذاری ایندکس رشته ها از 0 شروع میشود ینی از چپ به راست 
# شماره گذاری ایندکس رشته ها از آخر به اول ینی از راست به چپ با 1- شروع میشود
# برای مثال:
# "poria"
# 01234
# -5-4-3-2-1


name = "poria"
print(name[1])
# output: p

print(name[-1])
# output: a

print(name[0:3])
# output: por
# در این مثال در خروجی ایندکس 3 نمایش داده نمیشود و تا یکی آخری نمایش میدهد

print(name[0:5:2])
# output: pra
# در این مثال طول نمایش رشته و تعداد حرکت را مشخص کردیم

# برای نمایش برعکس رشته از آخر به اول میتوانیم به صورت زیر عمل کنیم:
print(name[::-1])
# output: airop
```

# متد های رشته ها

```python
# برای نمایش متد های هر نوع متغییری میتوانیم از تابع
# dir()
# استفاده کنیم

name = "poria"
print(dir(name))
# output:['__add__', '__class__', '__contains__', '__delattr__', ....

# تابع len
# برای نمایش تعداد عناصر داخل متغییر استفاده میشود
print(len(name))
# output: 5

# متد upper
print(name.upper())
# output: PORIA

# متد lower
name = "PORIA"
print(name.lower())
# output: poria


name = "poria"
# متد count
print(name.count("p"))
# output: 1
# نمایش تعداد کاراکتر ورودی متد

# متد endswith
print(name.endswith("a"))
# output: True
# این متد مشخص میکند رشته مورد نظر با ورودی متد شما به اتمام رسیده یا خیر

# متد startswith
print(name.startswith("p"))
# output: True
# این متد مشخص میکند رشته مورد نظر با ورودی متد شما شروع شده یا خیر

# متد find
print(name.find("p"))
# output: 0
# این متد شماره ایندکس ورودی متد را به شما میدهد

# متد rfind
print(name.rfind("i"))
# output: 0
# این متد شماره ایندکس ورودی متد را از سمت راست بررسی میکند و به شما میدهد

# متد isalnum
print(name.isalnum())
# output: True
# این متد مشخص میکند که تمام کاراکترهای رشته شما از نوع عددی و رشته ای است یا خیر
# تا وقتی از حروف و اعداد استفاده میشود True است

# متد isnumeric
print(name.isnumeric())
# output: False
# فقط زمانی درست است که تمام کاراکتر ها عددی باشد

# متد join
x = "*"
List = ["ali", "mmd", "reza"]
print(x.join(List))
# output: ali*mmd*reza

# متد splite
name = "poria, deli"
print(name.split(","))
# output: ["poria", "deli"]

# متد replace
name = "poria"
print(name.replace("p", "A"))
# output: Aoria

# متد strip
name = "    poria deli   "
print(name.strip())
# output: poria deli

# متد rstrip
name = "    poria deli   "
print(name.rstrip())
# output:     poria deli

# متد lstrip
name = "    poria deli   "
print(name.lstrip())
# output: poria deli    

# متد capitalize
name = "poria"
print(name.capitalize())
# output: Poria deli
```
