### سینتکس آرگومان ها

```python
# سینتکس Normal

def Max(a, b, c): # (پارامتر)
    print(max(a, b, c))

Max(9, 4, 3) # (آرگومان)
# output: 9

# درسینتکس نرمال:
# به ازای هر پارامتری که توی تعریف تابع نوشتی یک آرگومان ارسال کنید
# در این سینتکس باید حواستان به ترتیب باشد به این نوع سینتکس موقعیتی گفته میشود
# (a, b, c) پارامتر ها
# (9, 4, 3) آرگومان ها
```

```python
# name = value سینتکس

def Max(a, b, c): # (پارامتر)
    print(max(a, b, c))

Max(a=9, b=4, c=3) # (آرگومان)
# output: 9

# در این نوع سینتکس با قرار دادن اسم پارامتر مقدار آن را نیز با قرار دادن آرگومان مشخص کنید
# در این نوع سینتکس ترتیب مهم نیست
Max(b=4, c=3, a=9)
# output: 9
```

```python
# Normal + name=value سینتکس
def Max(a, b, c): # (پارامتر)
    print(max(a, b, c))

Max(9, 4, c=3) # (آرگومان)
# output: 9

# هم از آرگومان های موقعیتی و هم از آرگومان های نام و مقدار استفاده میشود
# اول ترتیب را مشخص میکنید و بعد نام مقدار
Max(9, c=3, b=4)
# output: 9
# یا
Max(9, b=4, c=3)
# output: 9

# اگر از این نوع سینتکس استفاده میکنید اول باید آرگومان هایی که براساس موقعیت هستند بنویسید
Max(2, 4, a=5) # سینتکس اررور
# output: Max() got multiple values for argument 'a'
```

```python
# *iterable سینتکس
# iterable: (set, list, str, tuple, ...)

def Func(a, b, c): # (پارامتر)
    print(f"a: {a}")
    print(f"b: {b}")
    print(f"c: {c}")

Func(*[4, 5, 9]) # (آرگومان)
# output: a: 4
#         b: 5
#         c: 9

# ------------------------------------------------------------------------------

# نکته: تعداد عناصر داخل ایتریبل ها باید با تعداد پارامتر ها یکی باشد نه کمتر نه بیشتر

# ------------------------------------------------------------------------------

Func(*"por")
# output: a: p
#         b: o
#         c: r

# ------------------------------------------------------------------------------

Func(*(2, 5, 8))
# output: a: 2
#         b: 5
#         c: 8

# ------------------------------------------------------------------------------

Func(*{7, 8, 9}) # ست ها چون ترتیب مهم نیست ممکنه ترتیب را بهم بزنند
# output: a: 8
#         b: 9
#         c: 7
```

```python
# **dict سینتکس

def Func(a, b, c): # (پارامتر)
    print(f"a: {a}")
    print(f"b: {b}")
    print(f"c: {c}")

d = {"a": 14, "b": 1, "c": 9} # کلید ها باید همان نام های پارامتر باشند ترتیب هم مهم نیست
Func(**d) # آرگومان
# output: a: 14
#         b: 1
#         c: 9
```
