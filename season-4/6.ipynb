# تابع Firstclass

#### این نوع تابع چه ویژگی هایی باید داشته باشد:

##### 1.can be created at runtime

```python
# تابعی که به شکل پویا بشه نابود یا ایجادش کرد
# برای مثال تابع را میتوانید داخل شرط حلقه ایجاد کنید
# برای مثال:

x = True
if x:
    def name():
        print("poria")
    name()
else:
    print("x not True")
# output: poria

# -----------------------------------------------------------------

x = False
if x:
    def name():
        print("poria")
    name()
else:
    print("x not True")
# output: x not True

# -----------------------------------------------------------------

# ایجاد توابع تو در تو
def A(a):
    print(a)
    
    def B(b):
        print(b ** 2)
    B(a) # داخل تابع اولی ایجاد و صدا میکنیم

A(5)
# output: 5
#         25
```

##### 2.can be assigned to a variable

```python
# توابع را میتوانید در یک متغییر ذخیره کنید
def Name(name):
    return f"My name is {name}"

new_name = Name # برای این کار باید تابع رو بدون پرانتز داخل متغییر قرار دهید
print(new_name("ali"))
# output: My name is ali
```

##### 3.can be passed as a argument to a function

```python
# توابع رو میتوانید به عنوان آرگومان در تابع دیگر بگیرید

def descending(mylist):
    print(sorted(mylist))


def ascending(mylist):
    print(sorted(mylist, reverse=True))

def MySort(Func, List): # آرگومان های تابع هر کدام از چپ به راست باید یک تابع و یک لیست باشد
    Func(List)

MySort(descending, [5, 9, 6, 3, 2, 1])
# output: [1, 2, 3, 5, 6, 9]
MySort(ascending, [5, 9, 6, 3, 2, 1])
# output: [9, 6, 5, 3, 2, 1]
```

##### 4.can be return as a result from a function

```python
# حروجی تابع میتواند یک تابع باشد

def MySort(String):
    
    def descending(mylist):
        print(sorted(mylist))
    
    def ascending(mylist):
        print(sorted(mylist, reverse=True))
    
    def Error(mylist):
        print("Error .....")
        print(mylist)
    
    if String == "d":
        return descending
    elif String == "a":
        return ascending
    else:
        return Error

action = input("Enter (d or a): ")

result = MySort(action)
result([5, 6, 9, 3, 1, 2, 4, 8])

# input: a
# output: [9, 8, 6, 5, 4, 3, 2, 1]

# input: d
# output: [1, 2, 3, 4, 5, 6, 8, 9]

# input: e
# output: Error .....
#         [5, 6, 9, 3, 1, 2, 4, 8]
```

##### 5.can have properties and methods

```python
# تابع ها میتوانند ی سری اتربیوت ها داشته باشند برای مثال:
def name():
    print("poria")

print(name.__name__)
# output: name

print(name.__class__)
# output: <class 'function'>

# برای دیدن تمامی موارد از تابع زیر استفاده کنید:
for atrr in dir(name):
    print(atrr)
# output: __annotations__
#         __builtins__
#         __call__
#         __class__
#         __closure__
#         __code__
#         __defaults__
#         __delattr__ 
#             .
#             .
#             .
#             .
```
