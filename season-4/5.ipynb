#### سینتکس docstring function

```python
def Max(x, y, z):
    """
    Receives three numbers as input and returns the largest of them.
    
    Parameters:
        x (int): A decimal integer
        y (int): Another decimal integer
        z (int): Another decimal integer
    
    Returns:
        max_int (int): largest of three numbers
    """
    return max(x, y, z)

print(Max.__doc__)
# output:  Receives three numbers as input and returns the largest of them.

#          Parameters:
#              x (int): A decimal integer
#              y (int): Another decimal integer
#              z (int): Another decimal integer

#          Returns:
#              max_int (int): largest of three numbers
```

#### سینتکس annotations

```python
def Max(a: int, b: int, c: int) -> int:
    return max(a, b, c)

print(Max(8, 9, 6))
# output: 9

print(Max.__annotations__)
# output: {'a': <class 'int'>, 'b': <class 'int'>, 'c': <class 'int'>, 'return': <class 'int'>}
```
