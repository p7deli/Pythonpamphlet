### سینتکس پارامترها

```python
# Normal سینتکس

def Func(a, b, c): # (پارامتر)
    print(f"a: {a}")
    print(f"b: {b}")
    print(f"c: {c}")

Func(2, 7, 8) # (آرگومان)
# outbut: a: 2
#         b: 7
#         c: 8
```

```python
# Default value
# در این نوع سینتکس شما چون مقدار دیفالت به پارامتر میدهید دیگر نیازی نیست آرگومان ارسال کنید
# اگر آرگومانی ارسال شود جاگزین مقدار دیفالت پارامتر میشود

def Func(a=2, b=4, c=3): # (پارامتر)
    print(f"a: {a}")
    print(f"b: {b}")
    print(f"c: {c}")

Func() # (آرگومان)
# outbut: a: 2
#         b: 4
#         c: 3

# ---------------------------------

Func(9) # (آرگومان)
# outbut: a: 9
#         b: 4
#         c: 3

# ---------------------------------

Func(9, 7, 6) # (آرگومان)
# outbut: a: 9
#         b: 7
#         c: 6

# ---------------------------------

Func(b=9) # (آرگومان)
# outbut: a: 2
#         b: 9
#         c: 3
```

```python
# Normal + Default value سینتکس
# در این نوع سینتکس شما هر مقدار دیفالتی که توی پارامتر تعیین کردی توی آرگومان میتونی ارسال نکنی
# آرگومان خالی نمیتونه باشه

def Func(a, b, c=3): # (پارامتر)
    print(f"a: {a}")
    print(f"b: {b}")
    print(f"c: {c}")

Func(9, 7) # (آرگومان)
# output: a: 9
#         b: 7
#         c: 3

# ---------------------------------

Func(9, 7, 6) # اگر پارامتر مقدار دیفالت بود و آرگومان ارسال شود مقدار آرگومان جایگزین مقدار پارامتر میشود
# output: a: 9
#         b: 7
#         c: 6

# ---------------------------------

Func(a=17, c=7, b=9)
# output: a: 17
#         b: 9
#         c: 7
```

```python
# *name سینتکس
# در این نوع سینتکس با قرار دادن کاراکتر استار پشت نام پارامتر بیشترین مقدار بعد از تریتب
# را برای خود بر میدارد

def Func(a, b, *c): # (پارامتر)
    print(f"a: {a}")
    print(f"b: {b}")
    print(f"c: {c}")

Func(9, 7, 14, 2, 8, 7) # (آرگومان)
# output: a: 9
#         b: 7
#         c: (14, 2, 8, 7)

# ---------------------------------

def Func(a, *b, c): # (پارامتر)
    print(f"a: {a}")
    print(f"b: {b}")
    print(f"c: {c}")

Func(9, 7, 14, 2, 8, c=7) # (آرگومان)
# output: a: 9
#         b: (7, 14, 2, 8)
#         c: 7

# ---------------------------------


def Func(*a, b, c): # (پارامتر)
    print(f"a: {a}")
    print(f"b: {b}")
    print(f"c: {c}")

Func(9, 7, 14, 2, b=8, c=7) # (آرگومان)
# output: a: (9, 7, 14, 2)
#         b: 8
#         c: 7
```

```python
# **name سینتکس

def Func(**a): # (پارامتر)
    print(f"a: {a['a']}")
    print(f"b: {a['b']}")
    print(f"c: {a['c']}")

Func(a=8, b=9, c=3)
# output: a: 8
#         b: 9
#         c: 3

# ---------------------------------

d = {"a": 14, "b": 4, "c": 7}
Func(**d) # (آرگومان)
# output: a: 14
#         b: 4
#         c: 7
```

```python
# استفاده از همه
def Func(a, **b): # (پارامتر)
    print(f"a: {a}")
    print(f"b: {b}")

Func(2, f=5, d=8, e=19) # آرگومان
# output: a: 2
#         b: {'f': 5, 'd': 8, 'e': 19}

# ---------------------------------

def Func(a, *b, **c): # (پارامتر)
    print(f"a: {a}")
    print(f"b: {b}")
    print(f"c: {c}")

Func(2, 5, 9, 7, f=5, d=8, e=19) # آرگومان
# output: a: 2
#         b: (5, 9, 7)
#         c: {'f': 5, 'd': 8, 'e': 19}
```
