#### 1. تابعی بنویسید که یک لیست از اعداد را به عنوان ورودی دریافت کند و بزرگترین عدد را برگرداند.

```python
# روش ساده 
def MaxList(List):
    return max(List)

print(f"max number in list: {MaxList([5, 6, 8, 3, 7])}")
# output: 8

# -----------------------------------------------------------

# روش پیچیده
def MaxList(List):
    max_number = 0
    for num in List:
        if num > max_number:
            max_number = num
    return max_number

print(f"max number in list: {MaxList([9, 6, 3, 5, 4, 2, 12])}")
# output: 12
```

#### 2. تابعی بنویسید که یک لیست از رشته‌ها را به عنوان ورودی دریافت کند و رشته‌ای با طول بیشترین کاراکتر را برگرداند.

```python
def MaxCharString(List):
    count, max_str = 0, ""
    for chr in List:
        if len(chr) > count:
            count, max_str = len(chr), chr
    return max_str

print(f"max string in list: {MaxCharString(['ali', 'poria', 'milad'])}")
```

#### 3. تابعی بنویسید که یک رشته و یک کاراکتر را به عنوان ورودی دریافت کند و تعداد تکرار آن کاراکتر در رشته را برگرداند.

```python
# روش ساده
def CountChar(String, Chr):
    return String.count(Chr)

St, Ch = input("Enter Sentence: "), input("Enter Chr: ")
print(f"{Ch} count in '{St}': {CountChar(St, Ch)}")
# input: my name is poria
# output: p count in 'my name is poria': 1

# -----------------------------------------------------------

# روش پیچیده
def CountChar(String, Char):
    counter = 0
    for chr in String:
        if chr == Char:
            counter += 1
    return counter

St, Ch = input("Enter Sentence: "), input("Enter Chr: ")
print(f"{Ch} count in '{St}': {CountChar(St, Ch)}")
# input: ali is a bad boy
# output: a count in 'ali is a bad boy': 3
```

#### 4. تابعی بنویسید که یک لیست از اعداد را به عنوان ورودی دریافت کند و مجموع اعداد آن را برگرداند.

```python
# روش ساده
def SumOfNumbers(List):
    return sum(List)

print(SumOfNumbers([4, 5, 6, 9, 8]))
# output: 32

# -----------------------------------------------------------

# روش پیچیده
def SumOfNumbers(List):
    Sum = 0
    for num in List:
        Sum += num
    return Sum

print(SumOfNumbers([5, 2, 0, 1, 7]))
# output: 15
```

#### 5. تابعی بنویسید که دو رشته را به عنوان ورودی دریافت کند و تعداد تکرار هر کاراکتر مشترک در دو رشته را برگرداند.

```python
def CountCommon(Str1, Str2):
    return len(set(Str1) & set(Str2))

print(CountCommon("poria", "ali"))
# output: 2
```

#### تابعی بنویسید که دو ورودی از کاربر بگیرد ورودی اول یک رشته و ورودی دوم یک جمله
#### و توی جمله جای تک رشته گرفته شده 
#### # قرار دهید

```python
def Change(String, Char):
    if len(Char) == 1: # شرط برای بررسی رشته تک عددی
        
        String = list(String) # تبدیل رشته به لیست برای تغییر پذیر کردن
        
        for chr in range(len(String)): # حرکت روی ایندکس عناصر لیست
            if String[chr] == Char: # شرط برای بررسی کاراکتر
                String[chr] = "#" # در صورت برقرار بودن شرط رشته مورد نظر با ایندکس تغییر میکند
        
        print(f"New Sentence: {''.join(String)}") # چاپ رشته تغییر داده شده و تبدیل آن از لیست به رشته
    else:
        print("Error...\nYour character input must be single letter") # در صورت برقرار نبودن شرط اول این متن چاپ میشود
    
Change("poria deli", "a")
# output: New Sentence: pori# deli

Change("my family is deli", "sa")
# output: Error...
#         Your character input must be single letter
```

#### تابعی بنویسید که یک ماتریس به عنوان ورودی دریافت کند و ماتریس را به صورت عمودی چاپ کند

```python
def print_vertical(matrix):
    for row in matrix:
        for item in row:
            print(item)
        print()

Matris = [[1, 2, 3],
          [8, 5, 9],
          [3, 6, 4]]
print_vertical(Matris)
# output: 1
#         2
#         3

#         8
#         5
#         9

#         3
#         6
#         4
```

#### تابعی بنویسید که یک لیست دریافت کند و اعداد زوج را از لیست پاک کند

```python
def DeleteEven(List):
    new_List = []
    for num in List:
        if num % 2 != 0:
            new_List.append(num)
    return new_List

print(DeleteEven([9, 6, 3, 5, 2, 4, 5, 8]))
# output: [9, 3, 5, 5]
```

#### تابعی بنویسید که یک لیست دریافت کند و عددی که بیشترین تکرار را دارد را برگرداند اگر همه اعداد یک بار تکرار شده باشند بزرگترین عدد لیست را برگرداند

```python
def Check(dic):
    for num in dic.values():
        if num >= 2:
            return False
    return True

def Biger(dic):
    number, count = 0, 0
    for num in dic.keys():
        if dic[num] > count:
            number, count = num, dic[num]
    
    return number, count

def repetition(List):
    Dic = {}
    for num in List:
        if num in Dic:
            Dic[num] += 1
        else:
            Dic[num] = 1
    
    if Check(Dic):
        return f"The largest number {max(List)}"
    else:
        res = Biger(Dic)
        return f"number: {res[0]}\ncounter: {res[1]}"

print(repetition([9, 6, 3, 8, 4]))
# output: The largest number 9
print(repetition([9, 6, 3, 8, 4, 4]))
# output: number: 4
#         counter: 2
```

#### تابعی بنویسید که یک رشته و یک کاراکتر به عنوان ورودی دریافت کرده و تعداد آن کاراکتر درون رشته را محاسبه کند و آن را برگرداند.

```python
def check_count_char(chr, String):
    counter = 0
    if chr in String:
        for char in String:
            if chr == char:
                counter += 1
        print(f"count {chr} in String: {counter}")
    else:
        print(f"Error ...\n{chr} not in String")
    
check_count_char("a", "hell my name is poria")
# output: count a in String: 2
check_count_char("g", "hell my name is poria")
# output: Error ...
#         g not in String
```

#### تابعی بنویسید که یک لیست از اعداد را به عنوان ورودی دریافت کرده و اعداد آن را به توان دو تبدیل کند و لیستی از اعداد به دست آورد که توان دو هر عدد در لیست قرار دارد و آن را برگرداند. 

```python
def pow_2(List):
    new_List = []
    for num in List:
        new_List.append(num ** 2)
    return new_List

print(pow_2([5, 6, 3, 2, 1, 4]))
# output: [25, 36, 9, 4, 1, 16]
```
