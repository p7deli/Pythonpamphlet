### توابع در پایتون یک بلوک کد هستند که یک نام دارند و می‌توانند توسط برنامه‌نویس تعریف شده و در هر زمانی که نیاز باشد فراخوانی شوند. تعریف تابع به صورت زیر است:

```python
#  - "def" کلید واژه‌ای است که برای تعریف یک تابع استفاده می‌شود.
#  - "نام_تابع" نام دلخواه برای تابع است که باید با حروف الفبا شروع شود.
#  - "پارامترها" لیستی از مقادیر
#  است که به عنوان ورودی به تابع ارسال می‌شود. این قسمت اختیاری است،
#  چنانچه تابع پارامتر ندارد، ممکن است این قسمت را خالی رها کنید.
#  - "بلوک_کد" قسمت اصلی تابع است که دستورات مورد نظر را دربرمی‌گیرد.
#  بلافاصله پس از تعریف تابع،
#  می‌توانید آن را با استفاده از نام آن وارد کنید و از آن استفاده کنید.
```

```python
# هدف اصلی از ایجاد تابع در پایتون، تجمیع یک گروه
# از دستورات و عملیات مشابه در یک بلاک کد است که می‌توان
# آن را به صورت تکراری استفاده کرد. با استفاده از توابع،
# می‌توان قسمت‌های مختلف برنامه را به صورت منظم و سازماندهی
# شده نگه داشت و قابل استفاده مجدد کرد. همچنین، توابع
# به برنامه نویسان این امکان را می‌دهند که قطعات کد خود
# را به صورت منطقی و قابل فهم برای خواننده نگارش کنند.
```

```python
# یک مثال ساده از ایجاد تابع
# f(x) = 2x + 1 فرمول

# برای شروع نام تابع را مشخص میکنیم و روردی آن را که پارامتر تابع نام دارد مشخص میکنیم
def f(x):
    # خروجی را به دو صورت میتوانیم نشان دهیم اولی:
    return 2 * x + 1

print(f(5))
# output: 11


def f(x):
    #دومی:
    print(2 * x + 1)

f(5)
# output: 11

# برای خروجی تابع طبق درخواست اگر از 
# return
# استفاده میکنید باید برای صدا کردن تابع و نشان دادن خروجی از 
# print()
# استفاده کنید ولی اگر خروجی تابع در داخل 
# print()
# استفاده شده برای نشان دادن خروجی نیازی نیست دوباره آن را داخل تابع 
# print()
# بزاریم
```
