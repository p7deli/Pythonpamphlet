## قوانین اسم‌گذاری تابع در پایتون عبارتند از:

##### 1. نام تابع باید با یک حرف یا underscore (_) شروع شود.
##### 2. نام تابع می‌تواند شامل حروف بزرگ و کوچک انگلیسی، اعداد و underscore باشد.
##### 3. نام تابع باید منحصر به فرد باشد و نباید با کلمات کلیدی (keywords) در پایتون تداخل داشته باشد.
##### 4. استفاده از حروف بزرگ در نام تابع به عنوان استاندارد پذیرفته شده نیست، به جای آن معمولاً از snake_case استفاده می‌شود که در آن کلمات با استفاده از underscore جدا می‌شوند.
##### 5. نام تابع باید معنای مناسب و قابل فهم داشته باشد و برای دیگران قابل خوانش و فهم باشد.
##### برخلاف زبان‌های دیگر، پایتون قوانین سخت و سریع خاص خود را در مورد طول نام تابع ندارد، اما به عموم معمولاً توصیه می‌شود که نام تابع به قدر کافی خلاصه و مفید باشد.

#### چند نمونه مثال برای مباحث تابع

```python
# برنامه ای بنویسید که دو عدد از کاربر بگیرد و نشان دهد اولی چند بار در دومی تکرار شده

def Repeat(digit, number):
    count = 0
    
    while number > 0:
        if (number % 10) == digit:
            count += 1
        number //= 10

    return count

num = int(input("Enter number: "))
numbers = int(input("Enter numbers: "))
count = Repeat(num, numbers)

print(f"count {num}: {count}")

# input: Enter number: 2
# input: Enter numbers: 265692425482
# output: count 2: 4
```

```python
# برنامه ای بنویسید که
# n فاکتوریل
# را محاسبه کند
# 1! + 2! + 3! + 4! + .... n!
# n = 3 => 1! + 2! + 3!:
#          (1) + (1 * 2) + (1 * 2 * 3): 1+2+6: 9

def Factoriel(number):
    Sub = 1
    for i in range(1, number+1):
        Sub *= i
    return Sub


def SumFactoriel(n):
    result = 0
    for num in range(1, n+1):
        result += Factoriel(num)
    return result


n = int(input("Enter n: "))
print(f"Factoriel {n}: {SumFactoriel(n)}")

# input: 3
# output: Factoriel 3: 9
```

```python
# برنامه ای بنویسید که 3 عدد از کاربر بگیرد و بزرگترین آن را چاپ کند

def MaxNumber(a, b, c):
    Max = max(a, b, c)
    print(Max)

MaxNumber(5, 9, 1)
# output: 9
```
