# سینتکس تابع

## قوانین اسم‌گذاری تابع در پایتون عبارتند از:

##### 1. نام تابع باید با یک حرف یا underscore (_) شروع شود.
##### 2. نام تابع می‌تواند شامل حروف بزرگ و کوچک انگلیسی، اعداد و underscore باشد.
##### 3. نام تابع باید منحصر به فرد باشد و نباید با کلمات کلیدی (keywords) در پایتون تداخل داشته باشد.
##### 4. استفاده از حروف بزرگ در نام تابع به عنوان استاندارد پذیرفته شده نیست، به جای آن معمولاً از snake_case استفاده می‌شود که در آن کلمات با استفاده از underscore جدا می‌شوند.
##### 5. نام تابع باید معنای مناسب و قابل فهم داشته باشد و برای دیگران قابل خوانش و فهم باشد.
##### برخلاف زبان‌های دیگر، پایتون قوانین سخت و سریع خاص خود را در مورد طول نام تابع ندارد، اما به عموم معمولاً توصیه می‌شود که نام تابع به قدر کافی خلاصه و مفید باشد.

```python
# توجه:
# تابع تا وقتی صدا زده نشن اجرا نمیشن

def name():
    print("poria")
# output: ....

# ------------------------------------------------------

def name():
    print("poria")
    
name()
# output: poria
```

```python
# خروجی تابع ها سلیقه ای است ینی یه وقتایی لازم داریم از خروجی تابع استفاده کنیم
# یه وقتایی هم لازم داریم که خروجی تابع را فقط نمایش دهیم
# مثال برای اولی:
def BakhshPazir(number):
    if number % 2 == 0:
        return True
    return False

result = BakhshPazir(8)
if result == True:
    print("Zoj")
else:
    print("Fard")

# output: Zoj

# -----------------------------------------------------------

# مثال برای دومی:
def BakhshPazir(number):
    if number % 2 == 0:
        print("Zoj")
    else:
        print("Fard")

BakhshPazir(9)
# output: Fard
```

```python
# کد هایی که بعد از 
# return
# نوشته میشن اجرا نمیشن و نادیده گرفته میشن

def name():
    return "poria"
    print("ali")
    print("ali")
    print("ali")

print(name())
# output: poria
```

```python
# تغییر نام تابع
def esm():
    print("poria")

name = esm # بدون پرانتز نام تابع قبلی را داخل متغییر جدیدی که ایجاد کردیم و میخواییم اسم جدید تابع باشدقرار میدهیم
name() # حالا اسم جدید تابع را با پرانتز صدا میزنیم تا خروجی نمایش داده شه
# output: poria
```
