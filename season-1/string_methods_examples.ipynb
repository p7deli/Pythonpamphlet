# center(width, fillchar)

##### وسط‌چین کردن رشته بین تعداد کاراکتر مشخص

```python
print("Hi".center(10, "-"))
```

## encodeencode(encoding)
### رشته را به بایت تبدیل می‌کند

```python
print("سلام".encode("utf-8"))
```

## expandtabs(tabsize)
#### تبدیل تب (\t) به فاصله

```python
text="A\tB\tC"
print(text.expandtabs(4))
```

## find(sub)
#### جستجو و برگرداندن اولین ایندکس (اگر نبود: -1)

```python
print("python".find("th"))
```

## format()
#### جایگذاری مقدار داخل متن

```python
print("Hello, {}".format("Pouria"))
```

## format_map(dict)
#### مثل format ولی با دیکشنری

```python
user={"name":"Pouria","age":22}
print("Name: {name}, Age: {age}".format_map(user))
```

## isascii()
#### چک می‌کند فقط کاراکترهای ASCII هستند؟

```python
print("hello123".isascii()); print("سلام".isascii())
```

## isdecimal()
#### فقط اعداد ده‌دهی؟

```python
print("123".isdecimal()); print("٣".isdecimal()); print("12.3".isdecimal())
```

## isdigit()
#### آیا فقط رقم است؟ شامل اعداد یونیکد مثل توان بالا و … هم می‌شود.

```python
print("123".isdigit()); print("²".isdigit())
```

## isidentifier()
#### آیا این متن یک نام متغیر معتبر در پایتون است؟

```python
print("user_name".isidentifier()); print("12name".isidentifier())
```

## isprintable()
#### آیا همه کاراکترها قابل چاپ هستند؟

```python
print("hello!".isprintable()); print("line\nbreak".isprintable())
```

## istitle()
#### آیا هر کلمه با حرف بزرگ شروع شده؟

```python
print("Hello World".istitle()); print("Hello world".istitle())
```

## ljust(width, fillchar)
#### چپ‌چین با پر کردن سمت راست

```python
print("Hi".ljust(10, "*"))
```

## maketrans / translate
#### ساخت جدول ترجمه

```python
table=str.maketrans({"a":"4","e":"3"})
print("apple".translate(table))
```

## removeprefix(prefix)
#### حذف پیشوند

```python
print("python_dev".removeprefix("python_"))
```

## removesuffix(suffix)
#### حذف پسوند

```python
print("file.txt".removesuffix(".txt"))
```

## rjust(width, fillchar)
#### راست‌چین با پر کردن سمت چپ

```python
print("Hi".rjust(10, "-"))
```

## translate(table)
#### جایگزینی کاراکترها با maketrans

```python
table=str.maketrans("abc","123")
print("abcabc".translate(table))
```

## zfill(width)
#### پر کردن سمت چپ با صفر

```python
print("42".zfill(5))
```
