کامنت‌ها در پایتون

کامنت‌ها در پایتون برای توضیح دادن کد به خواننده استفاده می‌شوند. کامنت‌ها با علامت # شروع می‌شوند و تا انتهای خط ادامه می‌یابند. هر چیزی که بعد از # نوشته شود، به عنوان کامنت در نظر گرفته می‌شود و توسط مفسر پایتون نادیده گرفته می‌شود.

مثال:

```python
# این یک کامنت است
print("poria")
```

داک استرینگ در پایتون

داک استرینگ یا Docstring در پایتون برای توضیح دادن توابع، کلاس‌ها و ماژول‌ها به کار می‌رود. داک استرینگ باید در ابتدای تابع، کلاس یا ماژول قرار گیرد و با """ شروع و با """ پایان یابد. داک استرینگ‌ها می‌توانند شامل توضیحات، مثال‌ها، پارامترها، نوع خروجی و ... باشند.

مثال:

```python
def func():
    """من یک داک استرینگ هستم"""
    print("Hello")

print(func.__doc__)
```

در نمونه مثال پایین شما میبینید که برای تعریف یک داک استرینگ نیاز نیست فقط از " استفاده کنید میتوانید از ' هم استفاده کنید

```python
def func():
    '''من یک داک استرینگ هستم'''
    print("Hello")

print(func.__doc__)
```

# و در آخر میپردازیم به تفاوت این دو

داک استرینگ ها و کامنت ها هر دو ابزارهای مفیدی برای توضیح دادن کد هستند، اما با هدف متفاوتی استفاده می شوند.

داک استرینگ ها یا مستندسازی کد، برای توضیح دادن عملکرد و استفاده از یک تابع، کلاس یا ماژول استفاده می شود. این ابزار به برنامه نویسان کمک می کند تا کد خود را بهتر فهمیده و به دیگران نیز امکان مشاهده و استفاده از آن را بدهد.

از طرفی، کامنت ها برای توضیح دادن جزئیات کد و یا اطلاعاتی که در داک استرینگ ها قرار نمی گیرند، مانند توضیحاتی در مورد چرایی استفاده از یک روش خاص، یا توضیحاتی در مورد نحوه عملکرد یک بخش خاص از کد، استفاده می شوند.

به طور کلی، داک استرینگ ها برای توضیح دادن کد به عنوان یک کل استفاده می شوند، در حالی که کامنت ها برای توضیح دادن جزئیات و جزئیات کد استفاده می شوند.
